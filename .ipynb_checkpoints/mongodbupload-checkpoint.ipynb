```python
import os
import certifi
import pymongo
import pandas as pd

# ✅ 1. Configuration
MONGO_DB_URL = "mongodb+srv://sanjaysahu121:Sahu2511@cluster0.3xnsxh0.mongodb.net/?appName=Cluster0"
DATABASE_NAME = "Spam_ham_detection" # Database name updated to match standard convention

# ✅ 2. Fixed Path for your local machine
# Yeh direct aapke error wale path ko target karega
DATASETS_DIR = r"c:\Users\91955\Machine Learning All\data"

# ✅ 3. Establish Connection
try:
    ca = certifi.where()
    client = pymongo.MongoClient(MONGO_DB_URL, tlsCAFile=ca)
    database = client[DATABASE_NAME]
    print("✅ Connected to MongoDB Atlas successfully")
except Exception as e:
    print(f"❌ Connection failed: {e}")
    exit()

# ✅ 4. Function to Upload ONLY spamham.csv
def upload_single_file_to_mongodb(datasets_dir):
    file = 'spamham.csv'
    file_path = os.path.join(datasets_dir, file)

    if not os.path.exists(file_path):
        print(f"❌ Error: File not found at {file_path}")
        return

    # Select the specific collection
    collection = database[file.split('.')[0]]
    print(f"📂 Processing: {file}")
    
    df = pd.read_csv(file_path)
    df.columns = df.columns.str.strip()
    data = df.to_dict(orient="records")

    if data:
        # Clear existing collection to avoid duplicates
        collection.delete_many({}) 
        
        # Upload in chunks of 5000 records to prevent timeouts
        chunk_size = 5000
        for i in range(0, len(data), chunk_size):
            collection.insert_many(data[i : i + chunk_size])
        
        print(f"✅ Success: {len(data)} records uploaded to '{file.split('.')[0]}'")
    else:
        print(f"⚠️ No data found in {file}")

# ✅ 5. Execution
if __name__ == "__main__":
    upload_single_file_to_mongodb(DATASETS_DIR)
```
